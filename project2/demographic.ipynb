```python
import pandas as pd
def demographic_data_analyzer(print_data=True):
    df = pd.read_csv('./adult.data',header=None,)
    df.columns=[
        'age', 'workclass', 'fnlwgt', 'education', 'education-num',
        'marital-status', 'occupation', 'relationship', 'race', 'sex',
        'capital-gain', 'capital-loss', 'hours-per-week', 'native-country', 'salary'
    ]
    race_count = df['race'].value_counts()
    if print_data:
        print("race_count=",race_count)

    df['education'] = df['education'].str.strip().str.lower()
    df['salary'] = df['salary'].str.strip()

    num_bachelors = len(df[df['education'].str.strip().str.lower() == 'bachelors'])
    print("number of people with a bachelor's degree:",num_bachelors)

    total_people = len(df)
    percentage_bachelors = (num_bachelors / total_people) *100
    print("percentage with bachelor degree=",round(percentage_bachelors,1),"%")

    higher_education = df[df['education'].isin(['bachelors','masters','doctorate'])]
    lower_education = df[~df['education'].isin(['bachelors','masters','doctorate'])]
    
    higher_education_rich = len(higher_education[higher_education['salary'] == '>50K'])
    lower_education_rich = len(lower_education[lower_education['salary'] == '>50K'])

    print("high edu earn >50k:",higher_education_rich)
    print("lower edu earn >50k:",lower_education_rich)

    higher_education_percentage = round((higher_education_rich/len(higher_education)) * 100, 1)
    lower_education_percentage = round((lower_education_rich/len(lower_education)) * 100, 1)

    print("percentage with advanced education earing >50k:",higher_education_percentage,"%")
    print("percentage without advanced education earing >50k:",lower_education_percentage,"%")

    min_work_hours = df['hours-per-week'].min()
    print("minimun work hours per week:",min_work_hours)

    min_workers = df[df['hours-per-week'] == min_work_hours]
    num_min_worker_rich = len(min_workers[min_workers['salary'] == '>50K'])
    print("number of worker who work minimun hours and earn >50k:",num_min_worker_rich)
    rich_percentage = round((num_min_worker_rich/len(min_workers)) *100,1)
    print("percentage of rich among those who work fewest hours:",rich_percentage,"%")

    country_total = df.groupby('native-country')['salary'].size()
    country_rich = df[df['salary'] == '>50K'].groupby('native-country').size()
    rich_percentage_by_country = (country_rich/country_total)*100 
    highest_earning_country = rich_percentage_by_country.idxmax()
    highest_earning_country_percentage = round((rich_percentage_by_country).max(),1)
    print("country with highest percentage of rich:",highest_earning_country)
    print("highest percentage of rich people:",highest_earning_country_percentage,"%")
    
    india_rich = df[(df['native-country'] == ' India') & (df['salary'] == '>50K')]
    top_In_occupation = india_rich['occupation'].value_counts().idxmax()
    print("most popular occupation for those who earn >50K in indai:",top_In_occupation)

demographic_data_analyzer()


```
